## üß† Was macht diese Funktion?

Die Funktion `ai_analyse_sentiment()` analysiert die **Stimmung eines deutschsprachigen Satzes** und ordnet diesen in **eine von f√ºnf Kategorien** ein:
`negativ`, `eher negativ`, `neutral`, `eher positiv`, `positiv` (f√ºnfstufige Likert-Skala).

Dabei nutzt die Funktion die Sprachf√§higkeiten des **GPT-4.1-mini-Modells**, um auch **implizite Ausdrucksweisen, Tonalit√§t und Konnotation** zu ber√ºcksichtigen ‚Äì also nicht nur offensichtliche W√∂rter, sondern auch Zwischent√∂ne.

Das Ergebnis eignet sich ideal f√ºr:

* qualitative oder quantitative UX-Auswertungen,
* automatisierte Klassifikation von Nutzerfeedback,
* didaktische Zwecke in der Sentimentanalyse.


## üîç Was wird bewertet?

Die Einsch√§tzung erfolgt rein sprachbasiert auf Grundlage eines einzelnen Satzes und ohne erkl√§renden Zusatz. GPT beurteilt:

* **Wortwahl und Ausdrucksweise**,
* **implizite Haltung** (z.‚ÄØB. Ironie, Frustration, Zustimmung),
* **emotionale Konnotation**.

Die R√ºckgabe erfolgt **immer exakt als ein Stringwert**, z.‚ÄØB. `"neutral"` oder `"eher positiv"` ‚Äì ohne Punkt, ohne Begr√ºndung.


## üß™ Beispiel aus der Praxis

Stell dir vor, du analysierst folgende Nutzerkommentare:

```python

# Einsch√§tzung "negativ"

"Mega gut! Gef√§llt mir echt."
# Einsch√§tzung "positiv"

"Danke."
# Einsch√§tzung "negativ"
```

## üéØ Warum ist das n√ºtzlich?

Diese Funktion bietet eine **einfache M√∂glichkeit, qualitative Daten zu strukturieren**. Gerade im UX-Kontext (z.‚ÄØB. Nutzerfeedback, Reviews, Support-Tickets) ist es hilfreich zu wissen:

* ob die R√ºckmeldung eher positiv oder negativ ist,
* wie stark die Bewertung gef√§rbt ist,
* ob es neutrale Aussagen gibt, die inhaltlich, aber nicht emotional gef√§rbt sind.

Auch im Unterrichtskontext kann die Funktion dazu genutzt werden, um **Textverarbeitung, Stimmungsanalyse und maschinelles Sprachverst√§ndnis zu vermitteln**.

* Wenn du **einzelne S√§tze** stimmungsm√§ssig einordnen m√∂chtest

**üí∏ Hinweis zu den Kosten**

Man k√∂nnte mehrere S√§tze in einem Prompt b√ºndeln, um API-Aufrufe zu sparen ‚Äì das w√ºrde jedoch **die Genauigkeit verringern** und **die Auswertung komplexer machen**.

üìå **Empfehlung**: F√ºr kleinere Datenmengen ideal ‚Äì bei vielen Texten lohnt sich ein Batch-Verfahren nur mit sorgf√§ltiger Validierung.


## ‚öôÔ∏è `ai_analyse_sentiment`

````python
from openai import OpenAI

# Setzen Sie hier Ihren OpenAI API-Key
OPENAI_API_KEY = "BITTE_F√úGE_DEINEN_EIGENEN_OPENAI_API-KEY_HIER_EIN"
GPT_MODEL = "gpt-4-turbo"

client = OpenAI(api_key=OPENAI_API_KEY)

def ai_analyse_sentiment(text: str) -> str:

    """
    F√ºhrt eine Sentimentanalyse einzelner deutschsprachiger S√§tze mit GPT durch.

    Die Funktion analysiert die allgemeine Tonalit√§t eines gegebenen Satzes anhand von Wortwahl, Ausdruck
    und impliziter Haltung. Dabei wird der Text vom Modell sprachlich interpretiert und einer von f√ºnf
    vordefinierten Stimmungskategorien zugeordnet: negativ, eher negativ, neutral, eher positiv oder positiv.
    Das zugrunde liegende Sprachmodell erkennt dabei auch feinere Nuancen der Sprache sowie indirekte Ausdrucksweisen.

    Parameter:
    - text: Ein einzelner Satz (String), dessen Stimmung eingesch√§tzt werden soll.

    R√ºckgabe:
    - Ein String mit genau einem der folgenden Werte:
        - 'negativ'
        - 'eher negativ'
        - 'neutral'
        - 'eher positiv'
        - 'positiv'

    Hinweis:
    - Die Analyse basiert auf dem OpenAI-Modell 'gpt-4.1-mini'.
    - Es wird keine Erkl√§rung oder zus√§tzliche Information zur√ºckgegeben ‚Äì nur die reine Kategorisierung.
    - Bei unklarer Modellantwort sollte ein Fehlerhandling erg√§nzt werden.
    """

    
    # Prompt
    prompt = f"""
    Analysiere die Stimmung des folgenden Satzes.
    Bewerte die allgemeine Tonalit√§t anhand der Wortwahl, Ausdrucksweise und impliziten Haltung.
    Gib ausschliesslich eine der folgenden f√ºnf Kategorien als Antwort zur√ºck (ohne zus√§tzliche Erkl√§rung):
    - negativ
    - eher negativ
    - neutral
    - eher positiv
    - positiv
    Satz: {text}
    """

    response = client.chat.completions.create(
        model=GPT_MODEL,
        messages=[
            {"role": "system", "content": (
                "Du bist ein zuverl√§ssiger Stimmungsanalysator. Deine Aufgabe ist es, die Stimmung einzelner S√§tze pr√§zise einzuordnen. "
                "Beurteile die Tonalit√§t jedes Satzes anhand von Wortwahl, Ausdruck, Implikation und Konnotation. "
                "Gib ausschliesslich **eine** der folgenden Kategorien als Antwort zur√ºck ‚Äì ohne Erkl√§rung, ohne zus√§tzliche W√∂rter, ohne Satzzeichen: "
                "negativ, eher negativ, neutral, eher positiv, positiv. Achte auf Konsistenz und vermeide Mehrdeutigkeiten. "
                "Antworte niemals mit etwas anderem als genau einem dieser Begriffe."
            )},
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )

    return response.choices[0].message.content.strip().lower()
````

## üöÄ Anwendungsbeispiel

````python
# einfaches Beispiel
text = "Das nicht so das, was ich erwartet habe."
result = ai_analyse_sentiment(text)
print(f"Sentiment: {result}")

text = "Ich bin total begeistert von diesem Service!"
result = ai_analyse_sentiment(text)
print(f"Sentiment: {result}")
````

````python
# Beispiel einer Tabelle mit mehreren Aussagen

import pandas as pd

# Beispiel-Daten: 5 S√§tze mit unterschiedlicher Stimmung
data = {
    "Kundenfeedbacks": [
        "Das nicht so das, was ich erwartet habe.",
        "Funktioniert ganz gut ‚Äì bin zufrieden.",
        "Ich bin total begeistert von diesem Service!",
        "Das Update war okay, aber nicht √ºberragend.",
        "Was f√ºr eine Entt√§uschung..."
    ]
}

# DataFrame erzeugen
df = pd.DataFrame(data)

# Funktion zur GPT-Sentimentanalyse (angenommen, bereits definiert)
df["Sentiment"] = df["Kundenfeedbacks"].apply(ai_analyse_sentiment)

# Ergebnis anzeigen
df
````

### Visualisierung

````python
import pandas as pd
import matplotlib.pyplot as plt

# Geordnete Sentiment-Kategorien
sentiments = ["positiv", "eher positiv", "neutral", "eher negativ", "negativ"]
data = df["Sentiment"].value_counts().reindex(sentiments, fill_value=0)

# Farbskala von gr√ºn (positiv) zu rot (negativ), modern ohne Warning
cmap = plt.colormaps.get_cmap("RdYlGn_r").resampled(len(data))

# Plot vorbereiten
fig, ax = plt.subplots(figsize=(8, 2), dpi=300)

left = 0
for i, (label, value) in enumerate(data.items()):
    color = cmap(i)
    ax.barh("Kundenfeedbacks", value, left=left, label=label, color=color)
    left += value

# Achsen & Layout
ax.set_xlim(0, data.sum())
ax.set_title("Verteilung der Sentiment-Klassifikationen")
ax.set_xlabel("Anzahl")
ax.set_ylabel("")
ax.legend(title="Sentiment", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Anzeige
plt.show()
````
